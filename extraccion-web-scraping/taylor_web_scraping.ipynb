# **Web Scraping de la tienda oficial de Taylor Swift**

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = 'https://store.taylorswift.com/collections/holiday-collection'
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.content,"html.parser")
```

```python
soup.body
```

```python
productos = soup.find_all("div", class_="ProductItem__Wrapper")
productos
```

###**Web Scraping - Holiday Collection**

```python
productos = soup.find_all("div", class_="ProductItem__Wrapper")
productos


productos_list = []
for p in productos:
  col = {}
  name = p.find("h2",class_="ProductItem__Title Heading").find('a').text
  price = p.find("div",class_="ProductItem__PriceList Heading")

  col['Nombre'] = name

  if price:
    col['Precio'] = price.find('span',class_="ProductItem__Price Price Text--subdued").text
  else:
    col['Precio'] = None

  productos_list.append(col)

holiday_coll_df = pd.DataFrame(productos_list)
holiday_coll_df
```

## Scraping Albums Shop

```python
# Funcion usando solo url
def get_taylor_shop(url):
  page = requests.get(url)
  soup = BeautifulSoup(page.content,"html.parser")

  productos = soup.find_all("div", class_="ProductItem__Wrapper")
  productos

  productos_list = []
  for p in productos:
    col = {}
    name = p.find("h2",class_="ProductItem__Title Heading").find('a').text
    price = p.find("div",class_="ProductItem__PriceList Heading")

    col['Nombre'] = name

    if price:
      col['Precio'] = price.find('span',class_="ProductItem__Price Price Text--subdued").text
    else:
      col['Precio'] = None

    productos_list.append(col)

  productos_df = pd.DataFrame(productos_list)
  return   productos_df
```

```python
# Taylor Swift Album Shop
ts_df = get_taylor_shop('https://store.taylorswift.com/collections/taylor-swift-album-shop')
ts_df
```

```python
# Fearless (Taylor's Version) Shop
fearless_df = get_taylor_shop('https://store.taylorswift.com/collections/fearless-taylors-version-shop')
fearless_df
```

```python
ts_df['Album'] = 'Taylor Swift'
fearless_df['Album'] = 'Fearless'
speak_now_df['Album'] = 'Speak Now'
red_df['Album'] = 'Red'
tv_df['Album'] = '1989'
reputation_df['Album'] = 'Reputation'
lover_df['Album'] = 'Lover'
folklore_df['Album'] = 'Folklore'
evermore_df['Album'] = 'Evermore'
midnights_df['Album'] = 'Midnights'
```

```python
shop_df = pd.concat([ts_df, fearless_df, speak_now_df, red_df, tv_df, reputation_df, lover_df, folklore_df, evermore_df, midnights_df], ignore_index=True)
shop_df
```

```python
shop_df.info()
```

```python
shop_df['Precio'] = shop_df['Precio'].str.replace('$', '')
shop_df['Precio'] = shop_df['Precio'].astype(float)
shop_df
```

```python
# Descarga
shop_df.to_csv('ws_taylor_store.csv', index=False)
```
